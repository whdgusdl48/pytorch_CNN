```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns

import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

from matplotlib.axes._axes import _log as matplotlib_axes_logger

# 아이리스 데이터셋 읽어오기!

iris = pd.read_csv("input/iris - iris.csv")
print(iris)
```

# DataSet 살펴보기

```python
iris["variety"].value_counts()
```

```python
iris.describe()
```

```python
iris.plot(kind="scatter",x="sepal.length",y='sepal.width',c = 'blue')
```

```python
sns.jointplot(x="sepal.length",y='sepal.width',data=iris, size=5)
```

```python
sns.FacetGrid(iris, hue="variety",size=5)\
    .map(plt.scatter, 'sepal.length','sepal.width')\
    .add_legend()
```

```python
sns.swarmplot(x='variety',y='petal.length',data=iris)
plt.grid()
```

```python
# boxplot
sns.boxplot(x='variety',y='petal.width',data=iris)
```

```python
# box plot
sns.boxplot(x='variety',y='sepal.width',data=iris)
```

```python
# boxplot by pandas
iris.drop('Id',axis=1).boxplot(by='variety',figsize=(12,6))
```

```python
# box plot + scatter
ax = sns.boxplot(x="variety",y="petal.length",data=iris)
ax = sns.stripplot(x='variety',y='petal.length',data=iris, jitter=True, edgecolor='gray')
```

```python
#violin plot

sns.violinplot(x='variety',y='petal.length',data=iris,size=6)
```

```python
# KDEPlot

sns.FacetGrid(iris,hue="variety",size=6)\
.map(sns.kdeplot,'petal.length',shade=True)\
.add_legend()
# petal.length
sns.FacetGrid(iris,hue="variety",size=6)\
.map(sns.kdeplot,'sepal.length',shade=True)\
.add_legend()
# sepal.length
sns.FacetGrid(iris,hue="variety",size=6)\
.map(sns.kdeplot,'petal.width',shade=True)\
.add_legend()
# petal.width
sns.FacetGrid(iris,hue="variety",size=6)\
.map(sns.kdeplot,'sepal.width',shade=True)\
.add_legend()
# sepal.width
```

```python
# pairplot
sns.pairplot(iris.drop("Id",axis=1),hue="variety",size=3,diag_kind="kde")
```

```python
# andrews plot

from pandas.plotting import andrews_curves
andrews_curves(iris.drop("Id",axis=1), 'variety')
```

```python
# parallel chart

from pandas.plotting import parallel_coordinates
parallel_coordinates(iris.drop("Id",axis=1), 'variety')
```
